```python
import pathlib
import json
```

```python
projRoot = pathlib.Path('..')
```

```python
public = projRoot / 'public'
audio = public / 'audio'
```

```python
sounds = [ {
        "name": path.stem,
        "audio": 'audio/'+str(path.name),
        "spectrogram": 'spectrograms/' + str(path.stem + ".png"),
        "img": ""
    } for path in audio.iterdir() ]
```

```python
sounds
```

```python
with (public / 'sounds.json').open('w') as f:
    json.dump(sounds, f, indent= 2)
```
